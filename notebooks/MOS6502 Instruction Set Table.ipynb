```python
import pandas as pd
rows = list(map(lambda x: x.split('\t'), table.strip().splitlines()))
df = pd.DataFrame(rows, columns=columns)
df.head()
```

```python
opcodes = df[df.Opcode != ''].dropna()
opcodes.head()
```

```python
opcodes.Mode.unique()
```

```python
opcodes['CPPMode'] = opcodes.Mode.map({
    'Absolute':      '__ABSOLUTE    ', 
    'Absolute, X':   '__ABSOLUTE_X  ', 
    'Absolute, Y':   '__ABSOLUTE_Y  ',    
    'Accumulator':   '__ACCUMULATOR ',
    'Immediate':     '__IMMEDIATE   ', 
    'Indirect':      '__INDIRECT    ', 
    '(Indirect, X)': '__INDIRECT_X  ', 
    '(Indirect), Y': '__INDIRECT_Y  ', 
    'Zero Page':     '__ZERO_PAGE   ', 
    'Zero Page, X':  '__ZERO_PAGE_X ', 
    'Zero Page, Y':  '__ZERO_PAGE_Y ',
    '':              '              ',
})
opcodes.head()
```

```python
opcodes['CPPHex'] = opcodes['Hex'].apply(lambda x: '0x' + x)
opcodes.head()
```

```python
opcodes['CPPenum'] = '\t' + opcodes['Opcode'] + opcodes['CPPMode'] + ' = ' + opcodes['CPPHex'] + ','
opcodes['CPPenum'].head()
```

```python
format_string = """enum class OpcodeTable: uint8_t {{
{}
}};"""
cpp = format_string.format('\n'.join(list(opcodes['CPPenum'].values)))
```
